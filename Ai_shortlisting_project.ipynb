```python
 total_required_skill = len(required_skills)
```

```python
def match_percentage(skills):
  matched = []
  for skill in required_skills:
    if skill in skills:
      matched.append(skill)
  return (len(matched)/total_required_skill)*100
```

```python
 df["match%"]= df["Skills"].apply(match_percentage)
```
